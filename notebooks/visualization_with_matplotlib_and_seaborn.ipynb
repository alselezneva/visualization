### Основные компоненты 

В библиотеке Matplotlib для Python, fig, ax и plt являются ключевыми компонентами для создания графиков. Вот их основные различия:

- **plt**: Это модуль в Matplotlib, который предоставляет функции для создания графиков. Он является удобным интерфейсом для рисования графиков с помощью простых команд. Например, plt.plot(x, y) используется для отображения графика линии между двумя наборами данных.
- **fig**: Это объект фигуры, который представляет собой окно или область, где будут отображаться графики. Вы можете думать о нем как о холсте, на котором вы рисуете свои графики. Вы можете изменять размеры фигуры, добавлять подписи к осям и многое другое.
- **ax**: Это объект осей, который представляет собой отдельную область внутри фигуры, где можно рисовать графики. Вы можете иметь несколько объектов ax в одной фигуре, что позволяет вам создавать сложные макеты с несколькими графиками. Каждый ax может иметь свои собственные оси и параметры.

### Разница между plt.plot() и ax.plot()

    plt.plot() 

```python
plt.figure(figsize=(10, 6), facecolor='red')

plt.plot(
    austin_weather["DATE"], 
    austin_weather["MLY-PRCP-NORMAL"], 
    label="Austin Monthly Precipitation",  # Добавляем легенду
    color="white",  # Изменяем цвет линии 
    linewidth=2,  # Увеличиваем ширину линии
    marker="o",  # Добавляем круглые маркеры к точкам
    markersize=5,  # Увеличиваем размер маркеров
    linestyle="--",  # Изменяем стиль линии на пунктирный
    alpha=0.7  # Уменьшаем прозрачность линии
)

plt.title("Monthly Precipitation in Austin")  # Добавляем заголовок графика
plt.xlabel("Date")  # Добавляем название оси X
plt.ylabel("Monthly Precipitation")  # Добавляем название оси Y
plt.legend()  # Отображаем легенду
plt.grid(True)  # Добавляем сетку

# Установка цвета фона осей 
plt.gca().set_facecolor('grey')

plt.show()
```

    ax.plot()

```python
fig, ax = plt.subplots(figsize=(10, 6), facecolor='red')

ax.plot(
    austin_weather["DATE"], 
    austin_weather["MLY-PRCP-NORMAL"], 
    label="Austin Monthly Precipitation",
    color="white",
    linewidth=2,
    marker="o",
    markersize=5,
    linestyle="--",
    alpha=0.7
)

ax.set_title("Monthly Precipitation in Austin")
ax.set_xlabel("Date")
ax.set_ylabel("Monthly Precipitation")
ax.legend()
ax.grid(True)
ax.set_facecolor('grey')

plt.show()
```

Принципиальная разница между использованием **plt.subplots()** и **plt.plot()** в matplotlib 
- **plt.subplots()**
Создает фигуру и набор осей (subplots) с одним вызовом.
Предоставляет контроль над тем, как создаются отдельные графики.
Рекомендуется для генерации нескольких подграфиков в сетках.
Может быть использован для создания пользовательских раскладок с помощью GridSpec или Figure.add_subplot для добавления подграфиков в произвольных местах внутри фигуры.
- **plt.plot()**
Используется для непосредственного рисования графика на текущей фигуре и оси.
Если только определена фигура, и вызывается plt.plot(), pyplot автоматически создаст оси на текущей фигуре и нарисует график на них.
Подходит для простых случаев, когда нужно быстро нарисовать график без необходимости создавать отдельные подграфики.

    Преимущество иметь figure - это возможность добавлять графики где угодно и когда угодно

```python
fig, ax = plt.subplots(figsize=(10, 6), facecolor='red')

ax.plot(
    austin_weather["DATE"], 
    austin_weather["MLY-PRCP-NORMAL"], 
    label="Austin Monthly Precipitation",
    color="white",
    linewidth=2,
    marker="o",
    markersize=5,
    linestyle="--",
    alpha=0.7
)

ax.set_title("Monthly Precipitation in Austin")
ax.set_xlabel("Date")
ax.set_ylabel("Monthly Precipitation")
ax.legend()
ax.grid(True)
ax.set_facecolor('grey')

add_ax1 = fig.add_axes([0, 0.6, 0.3, 0.3]) # left, bottom, wigth, height (range 0 to 1)
add_ax1.set_facecolor("blue")

add_ax2 = fig.add_axes([0.8, 0.2, 0.2, 0.4]) # left, bottom, wigth, height (range 0 to 1)
add_ax2.set_facecolor("m")

plt.show()
```

### Как еще можно настроить подграфики? 

    С помощью fig.axes - экзотично

```python
n = np.random.randn(1_000_000)

fig, _ = plt.subplots(1, 2, figsize=(16, 6))

fig.axes[0].hist(n, bins=100)
fig.axes[0].set_title("Default histogram")
fig.axes[0].set_xlim(min(n), max(n))                   

fig.axes[1].hist(n, cumulative=True, bins=500)  
fig.axes[1].set_title("Cumulative default histogram")  
fig.axes[1].set_xlim(min(n), max(n))   

plt.show()
```

**Использование fig.axes**:
fig.axes является списком всех объектов осей (axes) в фигуре. Вы можете обращаться к ним по индексу, начиная с 0. Этот подход подходит, когда вам нужно работать с несколькими объектами осей одновременно или когда вы хотите обращаться к оси по индексу, например, для циклической обработки всех осей в фигуре.


**Использование ax**:
ax обычно используется для обращения к конкретному объекту оси, который был возвращен функцией plt.subplots() или plt.subplot(). Этот подход предпочтительнее, когда вы работаете с одной осью за раз, так как он обеспечивает более прямой и понятный доступ к объекту оси.

Большой разницы нет, я советую использовать ax - я больше люблю этот способ, он как будто понятнее и привычнее :) 

    Ну и совсем *_магия_* с помощью GridSpec

```python
from matplotlib import gridspec
```

```python
fig = plt.figure(figsize=(10, 6))

# Создание сетки подграфиков
gs = gridspec.GridSpec(2, 6)  # Создание сетки из 2 строк и 6 столбцов
gs.update(wspace=1.5, hspace=0.3)  # Установка расстояния между столбцами и строками

# Создание подграфиков
ax1 = plt.subplot(gs[0, :2])  # Первый подграфик занимает первые 2 столбца первой строки
ax2 = plt.subplot(gs[0, 2:4])  # Второй подграфик занимает следующие 2 столбца первой строки
ax3 = plt.subplot(gs[0, 4:6])  # Третий подграфик занимает последние 2 столбца первой строки

ax4 = plt.subplot(gs[1, 1:3])  # Четвертый подграфик занимает первые 2 столбца второй строки
ax5 = plt.subplot(gs[1, 3:5])  # Пятый подграфик занимает следующие 2 столбца второй строки

# Добавление данных и настройки для каждого подграфика
# Например, добавление случайных данных для демонстрации
for ax in [ax1, ax2, ax3, ax4, ax5]:
    ax.plot(np.random.rand(10))

plt.show()
```

### Как построить графики в цикле? 

    1 - будет строить по столбцам (мне больше нравится именно этот способ, потому что здесь один for) 

```python
n_row = 2
n_col = 5

columns = list(austin_weather.columns[8:18])
assert len(columns) == n_row * n_col


fig, ax = plt.subplots(n_row, n_col, figsize=(30, 10))

for idx in range(len(columns)): 
    col = idx // n_row
    row = idx % n_row
    print(f'i:{idx}, col:{col}, row:{row}')

    ax[row,col].plot(
        austin_weather["DATE"], 
        austin_weather[columns[idx]] 
    )
    ax[row,col].set_title(f"idx:{idx}")
    
plt.show()
```

    2 - будет строить по строкам

```python
n_row = 2
n_col = 5

columns = list(austin_weather.columns[8:18])
assert len(columns) == n_row * n_col

fig, axes = plt.subplots(n_row, n_col, figsize=(30, 10))  

for row_idx, row_axes in enumerate(axes):
    for col_idx, ax in enumerate(row_axes):
        idx = row_idx * n_col + col_idx
        ax.plot(
            austin_weather["DATE"],
            austin_weather[columns[idx]] 
        )
        ax.set_title(f"idx:{idx}")
```

### А как нарисовать несколько линий на одном графике? 

    А это совсем просто !

```python
x = np.linspace(0, 5, 10)

fig, ax = plt.subplots(1, 2, figsize=(12, 5)) 

# можно прописать руками
ax[0].plot(x, x**2, label=f"y=x**2")
ax[0].plot(x, x**3, label=f"y=x**3")
ax[0].plot(x, x**4, label=f"y=x**4")
ax[0].legend(loc=2)

# или сделать цикл (это работает, если графики однотипные) 
[ax[1].plot(x, x**i, label=f"y=x**{i}") for i in [2, 3, 4]]
ax[1].legend(loc=2)

plt.show()
```

    На одном рисунке даже могут разместиться разные типы графиков ! (главное, чтобы это имело смысл + нужно аккуратно следить за осями) 

```python
x = np.linspace(0, 5, 10)

fig, ax = plt.subplots(1, 2, figsize=(12, 5)) 

# можно прописать руками
ax[0].scatter(x, x, label=f"y=x")
ax[0].hist(x,label=f"hist")
ax[0].legend(loc=2)

# или сделать цикл (это работает, если графики однотипные) - тут так просто уже не построить график 
[ax[1].plot(x, x**i, label=f"y=x**{i}") for i in [2, 3, 4]]
ax[1].legend(loc=2)

plt.show()
```

### Как можно передавать настройки для ax ? 

При использовании метода **set** для настройки графика в matplotlib, есть несколько ограничений и особенностей, которые следует учитывать:

- **Ограниченный набор параметров:** Метод set позволяет задать только определенный набор параметров, таких как xlim, ylim, xticks, yticks, xlabel, ylabel, title, и facecolor. Другие параметры, такие как linewidth, color, linestyle, и label, не могут быть установлены через set.

- **Непосредственное изменение параметров:** Параметры, установленные через set, изменяют значения непосредственно для текущего объекта Axes. Это означает, что если вы вызываете set для одного и того же объекта Axes несколько раз, последнее значение будет применено.

- **Совместимость с другими методами:** Некоторые параметры, которые можно установить через set, также могут быть установлены через отдельные методы, такие как set_xlim, set_ylim, set_xticks, set_yticks, и т.д. Это может привести к путанице, поскольку один и тот же параметр может быть установлен разными способами.

- **Не все параметры поддерживаются:** Некоторые параметры, которые можно установить через set, могут не поддерживаться в некоторых версиях matplotlib или для некоторых типов графиков. Например, параметр facecolor может не работать для некоторых объектов Axes.


```python
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# передали все параметры через set 
ax[0].plot(x, y, linewidth=5.0, color='r', linestyle='--', label='label')
ax[0].set(
    xlim=(-5, 5), 
    xticks=np.arange(1, 8),  
    ylim=(-1, 10), 
    yticks=np.arange(1, 8),
    xlabel='xlabel',
    ylabel='ylabel',
    title='title 1',
    facecolor='grey'
)
ax[0].legend()
ax[0].grid(False)

# передали каждый параметр через отдельный метод
ax[1].plot(x, y, linewidth=5.0, color='r', linestyle='--', label='label')
ax[1].set_xlim(-5, 5)
ax[1].set_ylim(-1, 10)
ax[1].set_xticks(np.arange(1, 8))  
ax[1].set_yticks(np.arange(1, 8))
ax[1].set_xlabel('xlabel')
ax[1].set_ylabel('ylabel')
ax[1].set_title('title 2')
ax[1].set_facecolor('grey')
ax[1].legend()
ax[1].grid(False)

fig.tight_layout()

plt.show()
```

    Сетка может быть основной (major) и дополнительной (minor). Каждому типу сетки можно задавать цвет, толщину линии и тип

```python
from matplotlib.ticker import AutoMinorLocator
```

```python
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)


fig, ax = plt.subplots(3, 2, figsize=(16, 10))

# -------------------------------------------------------------------------------------------------------------------------
# Первый график без дополнительных настроек 
ax[0,0].plot(x, y, linewidth=5.0, color='r', linestyle='--', label='label')
ax[0,0].grid(linewidth=1.2, color="black")


# -------------------------------------------------------------------------------------------------------------------------
# Второй график с настройками главных осей
x_ticks = np.arange(0, 17, 2)  # max(x)=10
y_ticks = np.arange(0, 11, 3)  
x_labels = [f"{i} label" for i in x_ticks]
y_labels = [f"{i} label" for i in y_ticks] 

ax[0,1].plot(x, y, linewidth=5.0, color='r', linestyle='--', label='label')
ax[0,1].grid(linewidth=1.2, color="black")
ax[0,1].set_xticks(ticks=x_ticks, labels=x_labels, rotation=45, horizontalalignment='right')
ax[0,1].set_yticks(ticks=y_ticks, labels=y_labels, rotation=45, horizontalalignment='right')

# эти параметры можно задавать отдельно друг от друга
# ax[0,1].set_xticks(x_ticks)
# ax[0,1].set_yticks(y_ticks)
# ax[0,1].set_xticklabels(x_labels, rotation=45, horizontalalignment='right')
# ax[0,1].set_yticklabels(y_labels, rotation=45, horizontalalignment='right')


# -------------------------------------------------------------------------------------------------------------------------
# Третий график с дополнительными осями
ax[1,0].plot(x, y, linewidth=5.0, color='r', linestyle='--', label='label')

# Установка дополнительных делений для осей X и Y
ax[1,0].xaxis.set_minor_locator(AutoMinorLocator()) # вертикальные линии - установка дополнительной сетки по оси X
ax[1,0].yaxis.set_minor_locator(AutoMinorLocator()) # горизонтальные линии - установка дополнительной сетки по оси Y

# Настройка параметров делений
ax[1,0].tick_params(which='major', length=10, width=5) # всем основным 
ax[1,0].tick_params(which='minor', length=5, width=1) # всем дополнительным 

# Настройка сетки
ax[1,0].grid(which="major", linewidth=1.2, color="black")
ax[1,0].grid(which="minor", linewidth=0.5, color="white")


# -------------------------------------------------------------------------------------------------------------------------
# Четвертый график с настройкой дополнительных осей 
ax[1,1].plot(x, y, linewidth=5.0, color='r', linestyle='--', label='label')

# Установка дополнительных (minor) делений для осей X и Y 
# Значение AutoMinorLocator должно быть больше 1 - означает, на сколько частей будем делить основной "квадратик" 
ax[1,1].xaxis.set_minor_locator(AutoMinorLocator(5)) # количество вертикальных линий - установка дополнительной сетки по оси X
ax[1,1].yaxis.set_minor_locator(AutoMinorLocator(10)) # количество горизонтальных линий - установка дополнительной сетки по оси Y

# Настройка параметров делений
ax[1,1].tick_params(which='major', length=10, width=5) # всем основным 
ax[1,1].tick_params(which='minor', length=5, width=1) # всем дополнительным 

# Настройка сетки
ax[1,1].grid(which="major", linewidth=1.2, color="black")
ax[1,1].grid(which="minor", linewidth=0.5, color="blue")

# Установка делений и меток для оси X и Y - именно они устанавливают major сетку 
x_ticks = np.arange(0, 17, 2)  # max(x)=10
y_ticks = np.arange(0, 11, 3)  
x_labels = [f"{i} label" for i in x_ticks]
y_labels = [f"{i} label" for i in y_ticks] 

ax[1,1].set_xticks(ticks=x_ticks, labels=x_labels, rotation=45, horizontalalignment='right')
ax[1,1].set_yticks(ticks=y_ticks, labels=y_labels, rotation=45, horizontalalignment='right')


# -------------------------------------------------------------------------------------------------------------------------
# Пятый график для экспериментов 
ax[2,0].plot(x, y, linewidth=5.0, color='r', linestyle='--', label='label')

# Установка дополнительных (minor) делений для осей X и Y 
# Значение AutoMinorLocator должно быть больше 1 - означает, на сколько частей будем делить основной "квадратик" 
ax[2,0].xaxis.set_minor_locator(AutoMinorLocator()) # количество вертикальных линий - установка дополнительной сетки по оси X
ax[2,0].yaxis.set_minor_locator(AutoMinorLocator()) # количество горизонтальных линий - установка дополнительной сетки по оси Y

# Настройка параметров делений
ax[2,0].tick_params(which='major', length=30, width=5) # всем основным 
ax[2,0].tick_params(which='minor', length=10, width=1) # всем дополнительным 

# Настройка сетки
ax[2,0].grid(which="major", linewidth=1.2, color="black")
ax[2,0].grid(which="minor", linewidth=0.5, color="white")

# Установка делений и меток для оси X и Y - именно они устанавливают major сетку 
x_ticks = np.arange(0, 17, 2)  # max(x)=10
y_ticks = np.arange(0, 11, 3)  
x_labels = [f"{i} label" for i in x_ticks]
y_labels = [f"{i} label" for i in y_ticks] 

ax[2,0].set_xticks(ticks=x_ticks, labels=x_labels, rotation=45, horizontalalignment='right')
ax[2,0].set_yticks(ticks=y_ticks, labels=y_labels, rotation=45, horizontalalignment='right')


# -------------------------------------------------------------------------------------------------------------------------
# Шестой график для экспериментов 
ax[2,1].plot(x, y, linewidth=5.0, color='r', linestyle='--', label='label')

# Установка дополнительных (minor) делений для осей X и Y 
# Значение AutoMinorLocator должно быть больше 1 - означает, на сколько частей будем делить основной "квадратик" 
ax[2,1].xaxis.set_minor_locator(AutoMinorLocator()) # количество вертикальных линий - установка дополнительной сетки по оси X
ax[2,1].yaxis.set_minor_locator(AutoMinorLocator()) # количество горизонтальных линий - установка дополнительной сетки по оси Y

# Настройка параметров делений
ax[2,1].tick_params(which='major', length=10, width=10) # всем основным 
ax[2,1].tick_params(which='minor', length=5, width=1) # всем дополнительным 

# Настройка сетки
ax[2,1].grid(which="major", linewidth=5, color="black")
ax[2,1].grid(which="minor", linewidth=0.5, color="white")

# Установка делений и меток для оси X и Y - именно они устанавливают major сетку 
x_ticks = np.arange(0, 17, 2)  # max(x)=10
y_ticks = np.arange(0, 11, 3)  
x_labels = [f"{i} label" for i in x_ticks]
y_labels = [f"{i} label" for i in y_ticks] 

ax[2,1].set_xticks(ticks=x_ticks, labels=x_labels, rotation=45, horizontalalignment='right')
ax[2,1].set_yticks(ticks=y_ticks, labels=y_labels, rotation=45, horizontalalignment='right')


fig.tight_layout()

plt.show()
```

### Основные типы графиков и их настройки 
https://matplotlib.org/stable/plot_types/index.html

**Pairwise data**
- plot(x, y)
- scatter(x, y)
- bar(x, height)
- stem(x, y)
- fill_between(x, y1, y2)
- stackplot(x, y)

**Statistical distributions**
- hist(x)
- boxplot(X)
- errorbar(x, y, yerr, xerr)
- violinplot(D)
- eventplot(D)
- hist2d(x, y)
- hexbin(x, y, C)
- pie(x)


#### plot + scatter plot + stacked plot

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)


fig, ax = plt.subplots(2, 3, figsize=(20, 10))  
 
# ------------------------------------------------------------------------------------
# plot 
ax[0,0].plot(x, y1, label='sin plot')
ax[0,0].plot(x, y2, linewidth=2.0, color='r', alpha=0.3, linestyle='--', label='cos plot')
ax[0,0].set_title("1 - plot")
ax[0,0].set_xlabel("Дата")
ax[0,0].set_ylabel("Объем осадков")
ax[0,0].legend(loc='lower left')

# fmt = '[marker][line][color]', аналогично через 'v--m' 
ax[1,0].plot(x, y1, linewidth=3, marker='v', linestyle='--', color='m', label='sin plot')
ax[1,0].plot(x, y2, 'b', linewidth=5.0, drawstyle='steps', label='cos plot')
ax[1,0].set_title("2 - plot")
ax[1,0].set_xlabel("Дата")
ax[1,0].set_ylabel("Объем осадков")

# ------------------------------------------------------------------------------------
# scatter 
ax[0,1].scatter(x, y1, color='m', marker='*', alpha=0.3, linewidths=5, label='sin plot')
ax[0,1].scatter(x, y2, label='cos plot')
ax[0,1].set_title("1 - scatter")
ax[0,1].set_xlabel("Дата")
ax[0,1].set_ylabel("Объем осадков")
ax[0,1].set_xlim(0, 10)
ax[0,1].set_ylim(-1, 1)

ax[1,1].scatter(x, y1, color='m', marker='*', alpha=0.3, linewidths=5, label='sin plot')
ax[1,1].scatter(x, y2, s=100, edgecolors='red', label='cos plot')
ax[1,1].set_title("2 - scatter")
ax[1,1].set_xlabel("Дата")
ax[1,1].set_ylabel("Объем осадков")
ax[1,1].set_xlim(0, 10)
ax[1,1].set_ylim(-1, 1)


# ------------------------------------------------------------------------------------
# stacked plot 
colors = ['blue', 'green', 'red']
labels=['sin(x)', 'cos(x)', 'cos(x)']

ax[0,2].stackplot(x, y1, y2, labels=labels, colors=colors)
ax[0,2].set_xlabel('X')
ax[0,2].set_ylabel('Y')
ax[0,2].set_title('1 - stacked plot')
ax[0,2].legend(loc='lower left')


ax[1,2].stackplot(x, y1, y2, y2, labels=labels, colors=colors, edgecolor='black', linewidth=2)
ax[1,2].set_title('2 - stacked plot')

fig.suptitle('Примеры для Plot', fontsize=30, fontweight='bold', verticalalignment='bottom')
fig.tight_layout()

plt.show()
```

```python
x = np.arange(1, 6)
y = np.random.rand(5)
yerr = np.random.rand(5)

categories = ['A', 'B', 'C', 'D', 'E']
values1 = [23, 45, 56, 33, 21]
values2 = [12, 34, 22, 17, 15]

fig, ax = plt.subplots(2, 2, figsize=(16, 10))  

# ------------------------------------------------------------------------------------
# bar 
ax[0,0].bar(categories, values1, width=1, label='Values 1')

bars = ax[1,0].bar(categories, values1, color='white', width=0.7, edgecolor='black', linewidth=3, label='Values 1')
for bar in bars:
    height = bar.get_height()
    ax[1,0].annotate(
        f'{height:.0f}',
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 3),  
        textcoords="offset points",
        ha='center', 
        va='bottom'
    )

# ------------------------------------------------------------------------------------
# errorbar  
ax[0,1].errorbar(x=categories, y=values1, yerr=yerr)
ax[0,1].bar(categories, values1, color='grey', alpha=0.4)
ax[0,1].set_title("Error Bar Plot")
ax[0,1].set_xlabel("X")
ax[0,1].set_ylabel("Y")

# ------------------------------------------------------------------------------------
# stacked bar

ax[1,1].bar(categories, values1, color='skyblue', alpha=0.7, label='Values 1')
ax[1,1].bar(categories, values2, bottom=values1, color='lightgreen', alpha=0.7, label='Values 2')
ax[1,1].bar(categories, values2, bottom=[v1+v2 for v1, v2 in zip(values1, values2)], color='red', alpha=0.4, label='Values 3')
ax[1,1].set_title("Stacked Bar Plot")
ax[1,1].legend(loc='upper right')
ax[1,1].set_ylim(0, 130)


fig.suptitle('Примеры для Bar', fontsize=30, fontweight='bold', verticalalignment='bottom')
fig.tight_layout()

plt.show()
```

#### boxplot 

Установка диапазона выбросов: whis позволяет установить диапазон выбросов, например, whis=(0, 100) или whis=[15, 85] для определенных процентилей.

```python
np.random.seed(19680801)
data = np.random.lognormal(size=(37, 4), mean=1.5, sigma=1.75)

# Настройки
boxprops = dict(linestyle='--', linewidth=3, color='darkgoldenrod')
flierprops = dict(marker='o', markerfacecolor='green', markersize=12, linestyle='none')
medianprops = dict(linestyle='-.', linewidth=2.5, color='firebrick')
meanpointprops = dict(marker='D', markeredgecolor='black', markerfacecolor='firebrick')
meanlineprops = dict(linestyle='--', linewidth=2.5, color='purple')


fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15, 8), sharey=True)

axs[0,0].boxplot(data, boxprops=boxprops, patch_artist=True)
axs[0,0].set_title('Custom boxprops', fontsize=10)

axs[0,1].boxplot(data, flierprops=flierprops, medianprops=medianprops, patch_artist=True)
axs[0,1].set_title('Custom medianprops\nand flierprops', fontsize=10)

axs[0,2].boxplot(data, whis=(0, 100), patch_artist=True)
axs[0,2].set_title('whis=(0, 100)', fontsize=10)

axs[1,0].boxplot(data, meanprops=meanpointprops, meanline=False, showmeans=True, patch_artist=True)
axs[1,0].set_title('Custom mean\nas point', fontsize=10)

axs[1,1].boxplot(data, meanprops=meanlineprops, meanline=True, showmeans=True, patch_artist=True)
axs[1,1].set_title('Custom mean\nas line', fontsize=10)

axs[1,2].boxplot(data, whis=[15, 85], patch_artist=True)
axs[1,2].set_title('whis=[15, 85]\n#percentiles', fontsize=10)


for ax in axs.flat:
    ax.set_yscale('log')
    ax.set_yticklabels([])

fig.suptitle("Box Plot Customization Examples")
fig.subplots_adjust(hspace=0.4)
plt.show()
```

#### Как поженить matplotlib и seaborn ?

#### seaborn.axisgrid.FacetGrid

**seaborn.axisgrid.FacetGrid** не является прямым аналогом **figure** в **matplotlib.pyplot**. Вместо этого, FacetGrid представляет собой объект, который позволяет создавать сложные сетки графиков, состоящие из множества подсубплотов, каждый из которых может быть индивидуально настроен. Это делает FacetGrid мощным инструментом для визуализации многомерных данных.

Отличия между FacetGrid и figure:
- **Функциональность:** FacetGrid предназначен для создания сеток графиков, где каждый подсубплот может иметь свои собственные оси и графики. Это отличает его от figure, который обычно используется для создания одиночного графика или группы графиков без внутренних сеток.
- **Возвращаемые значения:** Функции уровня figure в Seaborn, такие как sns.lmplot(), возвращают объект FacetGrid, который затем можно настроить с помощью методов FacetGrid. В отличие от этого, функции уровня axes в Seaborn, например, sns.countplot(), возвращают объект AxesSubplot из Matplotlib, который представляет собой отдельную область внутри figure для рисования графиков 2.
- **Использование:** FacetGrid идеально подходит для визуализации данных, когда необходимо сравнить категории по нескольким измерениям одновременно. Например, можно создать сетку графиков, где одна ось соответствует одному измерению, а другая — другому, и каждый подсубплот в сетке отображает третью переменную 1.
- **Настройка:** Для настройки графиков в FacetGrid используются специфические методы и атрибуты, такие как set_axis_labels(), set_titles(), despine(), и другие, которые не применимы к обычным объектам figure или Axes в Matplotlib 1.

В заключение, хотя FacetGrid и figure оба используются для создания графиков, они служат различным целям и имеют разные механизмы работы. FacetGrid предлагает более высокий уровень абстракции для создания сложных сеток графиков, в то время как figure предоставляет более низкоуровневый контроль над графиками и их расположением.

```python
g = sns.catplot()

print(type(g))
print(type(g.axes[0, 0]), end='\n\n') # это СЕТКА, поэтому надо 2 индекса !
print(' | '.join([i for i in dir(g) if (i[0] != '_')]), end='\n\n')
```
